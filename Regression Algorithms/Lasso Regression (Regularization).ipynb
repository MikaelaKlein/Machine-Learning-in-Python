```python
import numpy as np
from sklearn import linear_model
from sklearn.linear_model import Lasso
import sklearn.datasets
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
```

# Lasso Regression

Lasso Regression Equation:

\begin{align}
\hat{y} & = a_0 + a_1x_1  + a_2x_2 +  ...  + a_nx_n  + \alpha\sum_{i=1}^n \lvert a_i \rvert\\
\end{align}

This is the equation that the reduces the sum of squares the most. Choose a line that minimizes the error function. OLS (Ordinary Least Squares) minimizes the sum of squares of residuals.

Penalizes large coefficients because it leads to overfitting. Alpha is the coefficient of the error term, which is a coefficient we need to choose. This can be used to select important features of a dataset. It shrinks the coefficients of less important features to exactly 0.

## Example #1 - Boston Dataset

### Load the Data

```python
boston = sklearn.datasets.load_boston()

df_data = pd.DataFrame(boston.data, columns = boston.feature_names)
df_target = pd.DataFrame(boston.target, columns = ["MEDV"])
df = pd.concat([df_data, df_target], axis = 1)

X = df['RM'].values.reshape(-1, 1) # Only use one feature value, to be able to plot
y = df['MEDV'].values.reshape(-1, 1)
```

### Fit the Model

```python
lasso = Lasso(alpha = 0.1)
lasso.fit(X, y)
prediction_space = np.linspace(min(X), max(X)).reshape(-1, 1)
fig, ax = plt.subplots(figsize=(14, 8))
plt.scatter(X, y, color = 'blue')
plt.plot(prediction_space, lasso.predict(prediction_space), color = 'black', linewidth = 3)
plt.xlabel("Number of Rooms")
plt.ylabel("Values of house/1000 ($)")
```

### Model Evaluation

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
reg_all = Lasso(alpha = 0.1)
reg_all.fit(X_train, y_train)
y_pred = reg_all.predict(X_test)
reg_all.score(X_test, y_test)
```

### Cross-Validation

```python
cv_results = cross_val_score(lasso, X, y, cv = 5)
cv_results
```
