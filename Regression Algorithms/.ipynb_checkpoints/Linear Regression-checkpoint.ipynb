# Linear Regression

Linear Regression (OLS) Equation:

\begin{align}
\hat{y} & = \alpha_0 + \alpha_1x_1  + \alpha_2x_2 +  ...  + \alpha_nx_n \\
\end{align}

This is the equation that the reduces the sum of squares the most. Choose a line that minimizes the error function. OLS (Ordinary Least Squares) minimizes the sum of squares of residuals.

## Example #1 - Simple Linear Regression (Boston Dataset)

### Load the Data

```python
boston = sklearn.datasets.load_boston()

df_data = pd.DataFrame(boston.data, columns = boston.feature_names)
df_target = pd.DataFrame(boston.target, columns = ["MEDV"])
df = pd.concat([df_data, df_target], axis = 1)

X = df['RM'].values.reshape(-1, 1) # Only use one feature value, for simple linear regression
y = df['MEDV'].values.reshape(-1, 1)
```

### Fit the Model

```python
reg = linear_model.LinearRegression()
reg.fit(X, y)
prediction_space = np.linspace(min(X), max(X)).reshape(-1, 1)
fig, ax = plt.subplots(figsize=(14, 8))
plt.scatter(X, y, color = 'blue')
plt.plot(prediction_space, reg.predict(prediction_space), color = 'black', linewidth = 3)
plt.xlabel("Number of Rooms")
plt.ylabel("Values of house/1000 ($)")
```

### Model Evaluation

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
reg_all = linear_model.LinearRegression()
reg_all.fit(X_train, y_train)
y_pred = reg_all.predict(X_test)
reg_all.score(X_test, y_test)
```
